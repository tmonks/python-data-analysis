## Operations with Series

### Working at a column level, broadcast an update down the rows

```python
crisis = pd.Series([-1_000_000, -0.3], index=['GDP', 'HDI'])
crisis
```

```python
df[['GDP', 'HDI']]
```

```python
df[['GDP', 'HDI']] + crisis
```

### Set a column from a Series matching on index

```python
langs = pd.Series(['French', 'German', 'Italian'], index=['France', 'Germany', 'Italy'], name='language')
langs
```

Create a new column from the Series, where indexes match

```python
df['Language'] = langs
```

```python
df
```

You can also update an existing column in the same way

```python
langs['United States'] = 'English'
langs
```

```python
df['Language'] = langs
```

```python
df
```
