```python
import pandas as pd
```

```python
df = pd.read_csv('data/vgsales.csv', dtype={'Year': 'Int64'})
df.head()
```

```python
df.dtypes
```

## Group and count

```python
# Get the counts of games per Platform and Genre
df.groupby(['Platform', 'Genre'])['Platform'].size()
```

```python
# SQL style...
df.groupby(['Platform', 'Genre'], as_index=False)['Platform'].size()
```

## Pivot

Pivot and sum

```python
system_mask = df['Platform'].isin(['PS3', 'Wii', 'X360'])
df[system_mask].pivot_table(index='Year', columns='Platform', values='Global_Sales', aggfunc='sum', fill_value=0)
```
